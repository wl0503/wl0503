```python
import turtle

# Create a turtle screen and set up the turtle
screen = turtle.Screen()
t = turtle.Turtle()

# Define the ball properties
ballX = 100  # X-coordinate of the center of the ball
ballY = 100  # Y-coordinate of the center of the ball
ballSize = 50  # Radius of the ball

# Check if ballSize is greater than 0
if ballSize > 0:
    # Valid ballSize, draw the ball
    t.penup()
    t.goto(ballX, ballY - ballSize)
    t.pendown()
    t.begin_fill()
    t.circle(ballSize)
    t.end_fill()
else:
    # Invalid ballSize, handle the error or set a default size
    print("Invalid ballSize. Please use a positive value.")
    # Optionally, you can set a default size here
    # defaultBallSize = 10  # Set your default size
    # t.penup()
    # t.goto(ballX, ballY - defaultBallSize)
    # t.pendown()
    # t.begin_fill()
    # t.circle(defaultBallSize)
    # t.end_fill()

# Close the turtle graphics window when done
screen.exitonclick()
 
```

// Move the ball and check for collisions
function moveBall() {
  ballX += ballSpeedX;
  ballY += ballSpeedY;

  // Ball collision with top and bottom walls
  if (ballY < 0 || ballY > canvas.height) {
    ballSpeedY = -ballSpeedY;
  }

  // Ball collision with paddles
  if (
    ballX <= paddleWidth &&
    ballY >= leftPaddleY &&
    ballY <= leftPaddleY + paddleHeight
  ) {
    ballSpeedX = -ballSpeedX;
  } else if (
    ballX >= canvas.width - paddleWidth &&
    ballY >= rightPaddleY &&
    ballY <= rightPaddleY + paddleHeight
  ) {
    ballSpeedX = -ballSpeedX;
  }

  // Ball out of bounds (score)
  if (ballX < 0 || ballX > canvas.width) {
    // Reset the ball's position
    ballX = canvas.width / 2;
    ballY = canvas.height / 2;
    // Reset the ball's direction
    ballSpeedX = -ballSpeedX;
  }
}
